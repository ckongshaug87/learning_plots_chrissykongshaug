```python
import pandas as pd
import matplotlib.pyplot as plt
 
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
import numpy as p
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ["#7b92c9"] * len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()

highlight_position= avg_artist_streams.index.get_loc(max_index)

colors[highlight_position]= "#f030ad"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color =colors, width=.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout
plt.show()
```

```python
plt.Figure(figsize=(10,6))

special_artist = "Ariana Grande"
colors = ["#f030ad" if artist == special_artist else "gray" for artist in avg_artist_streams.index]


plt.bar(avg_artist_streams.index, avg_artist_streams.values, color =colors, width=.9)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout
plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#2bd808" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color=colors,
    alpha = 0.8,
    s = 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve HIgher Monthly Listens or Total Streams?", fontsize=18, pad=45)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(.5)

plt.tight_layout
plt.show()
```

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean()
yearly_avg
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    yearly_avg.index,
    yearly_avg.values,
    marker="+",
    linestyle="--",
    color="#20a32b",
    linewidth=0.5
)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad=20)

ax=plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color="#063b0a",
    linewidth= .5,
    alpha =.7
  
)

ax.text(
    2022+.05,
    3000,
    "Dip in Total Streams",
    fontsize=10
)


plt.ylim(0, 3500)
plt.show()
```

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python

fig, ax = plt.subplots(figsize=(8, 8))

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    labels=platform_totals.index,
    startangle=110,
    colors=["gray", "#1aaa10"]
)

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize=40,
    pad=20
)

for autotext in autotexts:
    autotext.set_fontsize(16)
    autotext.set_color("white")
    autotext.set_weight("bold")

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    title="Platform",
    bbox_to_anchor=(0.5, .95),
    ncol=2
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=15,
    color = "#2fba28",
    alpha=.5
)

plt.xlabel("Average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distribution of Average Stream Duration", fontsize=16, pad=20)

ax=plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(.5)
    spine.set_alpha(.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
plt.boxplot(
    [
        spotify_df[spotify_df["Platform Type"] == "Free"]["Skip Rate (%)"],
        spotify_df[spotify_df["Platform Type"] == "Premium"]["Skip Rate (%)"]
    ]
)

plt.xticks([1, 2], ["Free", "Premium"])
plt.xlabel("Platform Type")
plt.ylabel("Skip Rate (%)")
plt.title("Skip Rate % by Platform Type")

plt.show()
```

```python
import numpy as np
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

plt.figure(figsize=(10, 6))

im = plt.imshow(corr, vmin=-1, vmax=1, cmap="coolwarm")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)

plt.colorbar(im)

plt.title("Correlation Heatmap of Spotify Features")
plt.tight_layout()
plt.show()
```
